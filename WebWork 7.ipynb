```python
from collections import Counter
import string

def file2str (filename):
    with open(filename, encoding = "utf-8") as readthis: 
        returnString = readthis.read()
        
    return returnString

# Write a function init_counters that takes a list of strings -- words --
def init_counters(listOfStrings):
    # and initializes either a dict or a Counter to have values 0 for all keys.
    
    # here's how to initialize an empty counter
    word_counter = Counter()
    
    # here's how to pass it values
    word_counter = Counter(listOfStrings)
    
    return word_counter

def depunctify(text):
    #     Write a function called depunctify() that takes in a string, 
    #     and returns a new string that contains only alphanumeric characters and whitespace. 
    #     Be sure to deal with the double-hyphen problem; replace -- with a single space. 
    
    text = text.replace("--", " ")
    
    for punc in string.punctuation:
        text = text.replace(punc, " ")
    
    text = text.replace("’ ", " ")
    text = text.replace(" ’", " ")
    text = text.replace("‘", " ")
    
    return text

def canonicalize(text):
    #     de-punctifies it, (use your function from previous problem)
    #     replaces newlines with a space (why wouldn’t we just delete the newlines?), and
    #     de-capitalizes it.
    
    text = depunctify(text)
    
    text = text.replace("\n", " ")
    
    text = text.lower()
    
    return text

alice = file2str("Alice in Wonderland.txt")
test_words = ['dark','scary','power','light','happy','friend']
word_counter = init_counters(test_words)

alice_canonicalized = canonicalize(alice)

alice_list_words = alice_canonicalized.split()

len(alice_list_words)

alice_counter = init_counters(alice_list_words)

alice_counter = alice_counter.most_common()

alice_counter
```
